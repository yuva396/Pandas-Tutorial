```python
data = {
    'Name': ['Alice', 'bob', 'Charlie', np.nan, 'Eve'],
    'DOB': ['1998-05-14', np.nan, '1997-11-23', '2000-02-01', '1996-07-30'],
    'College': ['XYZ University', 'ABC College', np.nan, 'DEF College', 'LMN University']
}
```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.dropna()
```

```python
#only first one would be will be in df remanig all will be deleted

df.drop_duplicates(['Name'])
```

```python
#keep funcion helps to select which wants to be in df

df.drop_duplicates(['Name'], keep = 'last')
```

```python
df.drop_duplicates(['Name'], keep = False)
```

```python
# Example: Group by 'College' and count occurrences of names
df.groupby('College').size()
```

```python
# Example: Apply multiple aggregation functions to 'DOB' (e.g., count, first, last)
df.groupby('College')['DOB'].agg(['count', 'first', 'last'])
```

```python
# Concatenate two DataFrames
df2 = pd.DataFrame({
    'Name': ['Frank', 'Grace'],
    'DOB': ['2001-03-17', '2002-07-22'],
    'College': ['XYZ University', 'DEF College']
})

concatenated_df = pd.concat([df, df2], ignore_index=True)

# Merge DataFrames based on a common column
merged_df = pd.merge(df, df2, on='College', how='inner')
```

```python
merged_df
```

```python
# Convert all names to lowercase
df['Name'] = df['Name'].str.lower()

# Check if 'College' contains 'University'
df['College_contains_university'] = df['College'].str.contains('University')

# Replace part of a string
df['College'] = df['College'].str.replace('College', 'School')
```

```python
# Convert DOB to datetime format
df['DOB'] = pd.to_datetime(df['DOB'], errors='coerce')

# Extract the year from DOB
df['Year'] = df['DOB'].dt.year

# Extract the month from DOB
df['Month'] = df['DOB'].dt.month

# Handle missing dates using 'fillna' for default values
df['DOB'] = df['DOB'].fillna(pd.to_datetime('2000-01-01'))
```
